```python
def remove_char(this_char,row):
    that_row = row[0].split()
    new_row = []
    for char in that_row:
        if char != this_char:
            new_row.append(char)

    return new_row
```

```python
def process_row(row):
    row = remove_char("|",row)
    return row
    
```

```python
row_limit = 100
```

```python
def count_rows_and_columns(file_path): # still debugging, same code as the one below but with tqdm and progress bar 
    row_count = 0
    column_count = 0
    
    # First, get the total number of lines to initialize tqdm
    with open(file_path, mode='r', encoding='utf-8') as file:
        total_lines = sum(1 for line in file)
    
    # Now read the file again with tqdm progress bar
    with open(file_path, mode='r', encoding='utf-8') as file:
        for line in tqdm.tqdm(file, total=total_lines, desc="Processing rows"):
            row_count += 1
            if row_count == 1:
                column_count = len(line.split(','))

    return row_count, column_count

# count_rows_and_columns(big_df_path)
```

# A more advanced approach: splitting the dataset into train and test sets 

```python
def split_data(df,target_feature):
    new_df = df.copy()
    y = df[target_feature]
    x= df.drop(columns = [target_feature])

    return x,y
```

```python
x,y = split_data(table_df,'pressure_1')
```

```python
from sklearn.model_selection import train_test_split
```

```python
def split_train_test(df, target_feature, test_size = 0.2):
    df = df.copy()
    x , y = split_data(df,target_feature)
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = test_size, random_state = 42)
    return x_train, x_test, y_train, y_test
```
